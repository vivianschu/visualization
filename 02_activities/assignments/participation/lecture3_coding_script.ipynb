```python
# import libraries
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import scipy 
import PIL 
import requests

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
# plot line plot using both y1 and y2
ax.plot(x, y1)
ax.plot(x, y2)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
# plot line plot using both y1 and y2
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc='lower right')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
# plot line plot using both y1 and y2
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc='lower right',
          frameon=True, # add frame around legend
          fontsize=12,  # change font size
          ncol=2,   # specify num columns
          shadow=True)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
# plot line plot using both y1 and y2
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc='upper left',
          bbox_to_anchor = (1,1))
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
# plot line plot using both y1 and y2
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!")
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
# plot line plot using both y1 and y2
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!",
        ha='center',    # alignment of text
        color='red',    # modify font colour
        size=20)    # modify font size
fig.show()
```

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# arrow and annotation
ax.annotate('This is important!', xy=(10, 95), xytext=(20,94),
            arrowprops=dict(facecolor='black'))

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# different arrow style and colors
ax.annotate('This is important!', 
            xy=(10, 95), 
            xytext=(20,94),
            arrowprops=dict(arrowstyle="wedge",
                            color='hotpink'))

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# remove axis numbers
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# define max num ticks wanted on axis
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# set tick marks at particular interval (e.g., 5)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# rotating axis labels and horizontal alignment
plt.xticks(rotation=45, ha='right')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# activity: modifying axis title fonts
font1 = {'family': 'serif', 'color': 'indigo'}
plt.xlabel('Shiny New X Axis!', fontsize=18, fontdict=font1)
```
